Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Ans. Matplotlib is a widely-used Python library for creating static, animated, or interactive visualizations and plots. It provides a comprehensive set of tools for creating a wide range of charts, graphs, and plots for data analysis, data visualization, and presentation. Matplotlib is often used in data science, scientific research, and engineering fields.

Matplotlib is used for several purposes:

Data Visualization: It allows you to visualize data through various types of plots, charts, and graphs, making it easier to understand and interpret data.

Exploratory Data Analysis (EDA): It is an essential tool in the EDA process, helping data analysts and scientists explore data patterns and relationships.

Presentation: Matplotlib can be used to create publication-quality figures for research papers, reports, and presentations.

Customization: It offers a high degree of customization, allowing users to tailor the appearance and style of plots to suit their specific needs.

Interactivity: Matplotlib can be combined with libraries like Jupyter Notebook to create interactive visualizations for data exploration.


Five common plots that can be created using the Pyplot module of Matplotlib:

Line Plot: A basic line plot is used to display data points connected by straight lines. It is often used to show trends over time.

Scatter Plot: Scatter plots display individual data points as dots on a graph, making it easy to visualize the distribution and relationships between two variables.

Bar Plot: Bar plots are used to display categorical data with rectangular bars. They are often employed for comparing data among different categories.

Histogram: A histogram is used to represent the distribution of a dataset. It displays data in bins or intervals to show the frequency of values within each bin.

Pie Chart: Pie charts are circular representations of data, divided into slices to represent the relative proportions of different categories. They are useful for showing parts of a whole.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Ans.  A scatter plot is a type of data visualization that displays individual data points as dots (or markers) on a two-dimensional graph. It is used to visualize the distribution of data and the relationships between two variables. Scatter plots are particularly useful for identifying patterns, trends, and correlations in data.A scatter plot is a type of data visualization that displays individual data points as dots (or markers) on a two-dimensional graph. It is used to visualize the distribution of data and the relationships between two variables. Scatter plots are particularly useful for identifying patterns, trends, and correlations in data.

```python
import matplotlib.pyplot as plt
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random numbers")
plt.scatter(x,y)
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:
    
import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

Ans.

The subplot() function in Matplotlib is used to create multiple subplots within a single figure. This function allows you to arrange and display multiple plots in a grid or other customized layouts, making it easier to compare and visualize different aspects of your data in one figure.

The subplot() function takes three arguments: the number of rows, the number of columns, and the index of the current subplot. By specifying these arguments, you can create a grid of subplots and plot data in each of them.

```python

import numpy as np
import matplotlib.pyplot as plt

# Data for four line plots
x = np.array([0, 1, 2, 3, 4, 5])

# Create a figure with 2 rows and 2 columns
plt.figure(figsize=(10, 8))

# Plot Line 1
plt.subplot(2, 2, 1)
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.plot(x, y1)
plt.title('Line Plot 1')

# Plot Line 2
plt.subplot(2, 2, 2)
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.plot(x, y2)
plt.title('Line Plot 2')

# Plot Line 3
plt.subplot(2, 2, 3)
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.plot(x, y3)
plt.title('Line Plot 3')

# Plot Line 4
plt.subplot(2, 2, 4)
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.plot(x, y4)
plt.title('Line Plot 4')

# Adjust layout for better spacing
plt.tight_layout()

# Show the subplots
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

Ans. A bar plot, also known as a bar chart or bar graph, is a type of data visualization that uses rectangular bars to represent data. The length or height of each bar is proportional to the value it represents. Bar plots are used to compare and display the values of different categories or groups. They are particularly useful for showing the distribution of categorical data or comparing values across categories.

A horizontal bar plot is a bar plot in which the bars are oriented horizontally, making it suitable for displaying data where the categories or groups are better represented on the y-axis.

Bar plots are used for several purposes:

Comparing Categories: Bar plots are effective for comparing values or attributes of different categories.

Ranking and Sorting: They help in ranking and sorting categories based on a specific metric.

Visualizing Categorical Data: Bar plots are used to display and analyze categorical data.

Presenting Results: They are commonly used for presenting survey results, market data, and other categorical data sets.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.bar(company,profit)
plt.xlabel("Companies")
plt.ylabel("Profit")
plt.title("Company Profit")
plt.show()
```

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
plt.barh(company,profit)
plt.xlabel("Companies")
plt.ylabel("Profit")
plt.title("Company Profit")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

box1 = np.random.normal(100, 10, 200)

box2 = np.random.normal(90, 20, 200)

Ans.  

A box plot, also known as a box-and-whisker plot, is a type of data visualization that displays the distribution of a dataset. It is particularly useful for visualizing the summary statistics, central tendency, and spread of the data. Box plots are used to identify outliers, understand the skewness of the data, and compare distributions between different groups or categories.

box plots are used:

Identify Skewness: Box plots provide information about the skewness of the data. If the median is not at the center of the box, the data may be skewed.

Detect Outliers: Outliers are visible as individual data points beyond the whiskers, making them easy to spot.

Compare Distributions: Box plots are useful for comparing the distributions of multiple groups or categories.

Summary Statistics: They display key summary statistics, such as the median, quartiles, and range.

```python
import numpy as np
import matplotlib.pyplot as plt

box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


plt.boxplot([box1, box2], labels=['Box 1', 'Box 2'])
plt.xlabel('Value')
plt.title('Box Plot for Box 1 and Box 2')
plt.show()
```
