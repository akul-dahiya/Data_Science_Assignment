Q1. What is a database? Differentiate between SQL and NoSQL databases.

Ans. A database is a structured collection of data that is organized, stored, and managed to facilitate data retrieval, storage, and manipulation. Databases are essential for efficiently storing and accessing data in a way that allows for easy retrieval and analysis. 
Databases typically consist of tables, where each table contains rows and columns, similar to a spreadsheet. The data is stored in a structured format, allowing for efficient querying, searching, and updating of information.

Difference between SQL and NoSQL databases is as following:
 SQL databases are relational databases, meaning they use structured schemas and tables to organize data. Data is stored in rows and columns, and relationships between tables are established through foreign keys.NoSQL databases are non-relational databases, which means they do not use structured schemas. They can store data in various formats, such as key-value pairs, document stores, wide-column stores, or graph databases.
SQL databases have a fixed schema, which means the structure of the data (the tables, columns, and data types) is defined before data insertion. Any changes to the schema require alterations to the existing structure. NoSQL databases are schema-less, allowing for flexibility in data storage. You can insert data without defining a fixed schema in advance.  SQL databases use SQL, a standardized query language, for data manipulation and retrieval. SQL is powerful for complex queries and joins. NoSQL databases have their query languages, which are often less powerful than SQL but tailored to specific data models. Some NoSQL databases support SQL-like querying.


Q2. What is DDL? Explain why CREATE, DROP, ALTER, and TRUNCATE are used with an example.

Ans.  DDL(Data Defination Language) is a subset of SQL (Structured Query Language) that is used to define, manage, and manipulate the structure and schema of a database. DDL commands are responsible for creating, modifying, and deleting database objects, such as tables, indexes, and constraints. These commands enable database administrators and developers to define how data is organized within the database.


CREATE:

The CREATE statement is used to create new database objects, such as tables, indexes, or views.
Example:

CREATE TABLE Employees (

    EmployeeID INT PRIMARY KEY ,
    
    FirstName VARCHAR(50) ,
    
    LastName VARCHAR(50),
     
    PhoneNumber INT(10)
);

In this example we are creating a table name Employess and different column

DROP:

The DROP statement is used to delete existing database objects, such as tables or indexes. It permanently removes the object from the database.

Example:

DROP TABLE Employees;

ALTER:

The ALTER statement is used to modify the structure of an existing database object. It can be used to add, modify, or delete columns in a table, among other things.

Example:

ALTER TABLE Employees

ADD Salary DECIMAL(10, 2);

in this example adding a new column of salary in employees table

TRUNCATE:

The TRUNCATE statement is used to quickly delete all rows from a table while keeping the table structure intact. It is faster and more efficient than the DELETE statement when you want to remove all data from a table.

Example:

TRUNCATE TABLE Employees;




Q3. What is DML? Explain INSERT, UPDATE, and DELETE with an example.

Ans. DML(Data Manipulation Language) is a subset of SQL (Structured Query Language) used to interact with and manipulate data stored in a database. DML commands are responsible for inserting, updating, retrieving, and deleting data in database tables. DML operations allow users and applications to modify the contents of a database, ensuring that data remains accurate and up-to-date.

INSERT:

The INSERT statement is used to add new records (rows) into a table.

Example:

INSERT INTO Employees (EmployeeID, FirstName, LastName, PhoneNumber)

VALUES (101, 'John', 'Doe', 9484753722);

UPDATE:

The UPDATE statement is used to modify existing records in a table. It allows you to change the values of specific columns in one or more rows based on specified conditions.

Example:

UPDATE Employees

SET LastName = 'Kumar'

WHERE EmployeeID = 101;

DELETE:

The DELETE statement is used to remove one or more records from a table based on specified conditions.

Example:

DELETE FROM Employees

WHERE EmployeeID = 101;



Q4. What is DQL? Explain SELECT with an example.

The result of a SELECT query is a result set, which is a set of rows that match the specified criteria.

Example:

Q5. Explain Primary Key and Foreign Key.

Ans.

Primary Key :

A Primary Key is a field or a combination of fields in a database table that uniquely identifies each record (row) within that table.
The primary key enforces entity integrity, which means that each row in the table must have a unique identifier, and no two rows can have the same primary key value.
Typically, primary keys are used to establish relationships between tables, ensuring that related data can be linked correctly.
Primary keys can consist of one or more columns (composite primary key), depending on the database design.
It's important to note that primary key values cannot contain NULL values.

Foreign Key :

A Foreign Key is a field in a database table that is used to establish a link between the data in two tables.
The foreign key creates a relationship between tables by referencing the primary key of another table. It enforces referential integrity, ensuring that data in the related tables remains consistent.
Foreign keys help maintain data integrity by preventing actions that would destroy links between related data. They typically constrain the values in the foreign key column(s) to match values in the referenced primary key column(s).
Foreign keys are used to create relationships between tables, such as one-to-many or many-to-many relationships.

Q7. Give the order of execution of SQL clauses in an SQL query.

Ans. In an SQL query, the SQL clauses are executed in a specific order, which is important to understand to correctly write and interpret SQL statements. The typical order of execution for SQL clauses in a query is as follows:

FROM Clause: The FROM clause is the first clause to be executed. It specifies the tables or data sources from which the data will be retrieved. It can also include joins to combine data from multiple tables.

JOIN Clause: If there are any join operations specified in the FROM clause, they are executed next. Joins are used to combine data from multiple tables based on specified conditions.

WHERE Clause: The WHERE clause is executed after the FROM and JOIN clauses. It filters the rows from the tables that meet the specified conditions. Rows that do not satisfy the conditions are excluded from the result set.

GROUP BY Clause: If a GROUP BY clause is present, it is executed next. It groups the rows in the result set based on one or more columns. Aggregation functions like SUM, COUNT, AVG, etc., can be applied to these groups.

HAVING Clause: If a HAVING clause is used in conjunction with GROUP BY, it filters the groups created by the GROUP BY clause based on specified conditions.

SELECT Clause: The SELECT clause determines which columns are included in the final result set. It specifies what data is retrieved from the tables or grouped data.

DISTINCT Clause: If the DISTINCT keyword is used in the SELECT clause, it eliminates duplicate rows from the result set.

ORDER BY Clause: The ORDER BY clause is executed last. It sorts the rows in the result set based on one or more columns in ascending or descending order.

LIMIT/OFFSET Clause (if supported): Some database systems support the LIMIT and OFFSET clauses, which can be used for pagination or limiting the number of rows returned in the result set. These clauses are executed after the sorting specified in the ORDER BY clause.

Result: Finally, the result set that matches the specified criteria and order is returned
